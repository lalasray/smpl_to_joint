```python
import numpy as np
from scipy.spatial.transform import Rotation as R

def integrate_angular_velocity(
        omega,                    # (N, 3)  rad/s
        dt,
        init_quat=np.array([0, 0, 0, 1]),  # [x,y,z,w]  worldbody (or identity)
        world_coordinates=False    # True if  is expressed in world frame
):
    """
    From a sequence of angular-velocity samples returns the corresponding
    orientation quaternions q[k] (length N+1).

    Returns: (N+1, 4) quaternions, scalar-last [x,y,z,w]
    """
    N          = omega.shape[0]
    q_out      = np.empty((N + 1, 4))
    q_curr     = R.from_quat(init_quat)

    q_out[0] = init_quat
    for k in range(N):
        # incremental rotation over this step
        dq = R.from_rotvec(omega[k] * dt)     # world axisangle

        # accumulate
        q_curr = dq * q_curr if world_coordinates else q_curr * dq
        q_out[k + 1] = q_curr.as_quat()

    print(q_out.shape)

    return q_out



```
